```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['font.family'] = 'Helvetica'
units = 'weeks'
model = 'contemp'

include_ksu = True
```

```python
#weeks_results_dir = '../results/assoc_contemp_lagged_weeks_2024-05-11_18_47_40.615096/'
#months_results_dir = '../results/assoc_contemp_lagged_months_2024-05-11_18_44_16.131608/'

weeks_results_dir = '../results/assoc_contemp_lagged_2024_weeks_2024-06-26_15_47_00.166579/'
months_results_dir = '../results/assoc_contemp_lagged_2024_months_2024-06-26_15_35_52.478345/'

df_months = pd.read_csv(months_results_dir + model + '_df_months_all_primary.csv').rename(columns={'Unnamed: 0': 'category'})

df_weeks = pd.read_csv(weeks_results_dir + model + '_df_weeks_all_primary.csv').rename(columns={'Unnamed: 0': 'category'})

annot_df_weeks = pd.read_csv(weeks_results_dir +'annot_' + model + '_df_weeks_all_primary.csv').rename(columns={'Unnamed: 0': 'category'})

annot_df_months = pd.read_csv(months_results_dir +'annot_' + model + '_df_months_all_primary.csv').rename(columns={'Unnamed: 0': 'category'})

#lagged_df_months
ksu_df = df_months.reindex([3,4,5])

ksu_df['Okja_se'] = ksu_df['Okja/WTH_se']
ksu_df['WTH_se'] = ksu_df['Okja/WTH_se']
ksu_df['Okja_pe'] = ksu_df['Okja/WTH_pe']
ksu_df['WTH_pe'] = ksu_df['Okja/WTH_pe']
ksu_df = ksu_df.drop(['Okja/WTH_se', 'Okja/WTH_pe'],axis=1)
```

```python
for col in annot_df_weeks.columns:
    if col == 'category':
        continue
    annot_df_weeks[col + '_an'] = annot_df_weeks[col]
    annot_df_weeks = annot_df_weeks.drop([col],axis=1)
```

```python
for col in annot_df_months.columns:
    if col == 'category':
        continue
    annot_df_months[col + '_an'] = annot_df_months[col]
    annot_df_months = annot_df_months.drop([col],axis=1)
```

```python
if units == 'weeks':
    df_weeks = pd.concat([df_weeks, ksu_df]).reset_index(drop=True)
    
```

```python
#reordering = [0, 9, 12, 1, 7, 10, 2, 8, 11, 3, 4, 6]
if units == 'weeks':
    reordering = [0,1,2,3,4,5,12,13,14,8, 11, 6, 9, 7, 10]#[9, 12, 7, 10, 8, 11, 3, 4, 6]
    df = df_weeks.reindex(reordering).reset_index(drop=True).rename(columns={'Outcome': 'category'})
    
    # Create merged annot_df
    meat_demand_categories = ['Beef Demand', 'Pork Demand', 'Chicken Demand']
    annot_df_months_meat = annot_df_months.query('category in @meat_demand_categories')
    
    annot_df_weeks = pd.concat([annot_df_weeks, annot_df_months_meat])
    
    df = df.merge(annot_df_weeks, on='category', how='inner')
    
elif units == 'months':
    reordering = [0, 1, 2, 6, 7, 8, 3, 4, 5,  11, 14, 9, 12, 10, 13]
    
    df = df_months.reindex(reordering).reset_index(drop=True).rename(columns={'Outcome': 'category'})

    df = df.merge(annot_df_months, on='category', how='inner')
```

```python
df['category'] = df['category'].map({'StewartMilk': 'Milk Purchase', 'StewartPBMilk': 'Plant-Based Milk Purchase', 'Zhao': 'Plant-Based Meat Market Share'}).fillna(df['category'])
```

```python
df.loc[df['category'] == 'Beef Demand'].filter(regex='Cowspiracy')
```

```python
# Drop NeuhoferLusk
df = df.loc[df['category'] != 'NeuhoferLusk'].reset_index(drop=True) 
if not include_ksu:
    df = df.loc[df['category'] != 'Beef Demand'].reset_index(drop=True)      
    df = df.loc[df['category'] != 'Pork Demand'].reset_index(drop=True)      
    df = df.loc[df['category'] != 'Chicken Demand'].reset_index(drop=True)      
```

```python
#intvns = ['FOK', 'Cowspiracy', 'Okja/WTH', 'TGC', 'All']
#intvns = ['FOK', 'Cowspiracy', 'Okja', 'WTH', 'TGC', 'All']

if units == 'weeks':
    intvns = ['FOK', 'Cowspiracy', 'WTH', 'Okja', 'TGC', 'YAWYE']

elif units == 'months':
    intvns = ['FOK', 'Cowspiracy', 'Okja/WTH', 'TGC', 'YAWYE']
    
    
for intvn in intvns:
    df[intvn + '_ci_lower'] = df[intvn + '_pe'] - 1.96*df[intvn + '_se'] 
    df[intvn + '_ci_upper'] = df[intvn + '_pe'] + 1.96*df[intvn + '_se'] 
```

```python
df = df.reindex(range(df.shape[0])[::-1])
data = df
```

```python
outcomes = list(data['category'])
interventions = intvns
if units == 'weeks':
    colors = ['blue', 'orange', 'green', 'red', 'purple', 'black']
else:
    colors = ['blue', 'orange', 'green', 'purple', 'black']
```

```python
intvn_map = {'FOK': 'Forks Over Knives', 'TGC': 'The Game Changers',
            'Okja': 'Okja', 'Cowspiracy': 'Cowspiracy', 'WTH': 'What the Health',
            'Okja/WTH': 'What the Health / Okja', 'YAWYE': 'You Are What You Eat'}
```

```python
current_pos = 0
increment = 0.1
between_outcome_increment = 0.5

fig, ax = plt.subplots(figsize=(10,10))


for i, outcome in enumerate(outcomes):
    for j, intervention in enumerate(interventions):
        pe = list(data[intervention + '_pe'])[i] #data[outcome][intervention]['pe']
        ci_lower = list(data[intervention + '_ci_lower'])[i] #data[outcome][intervention]['ci_lower']
        ci_upper = list(data[intervention + '_ci_upper'])[i]#data[outcome][intervention]['ci_upper']
        xerr_lower = pe - ci_lower
        xerr_upper = ci_upper - pe
        
        if pe == 0.00000:
            if j == (len(interventions) - 1):
                #plt.axhline(current_pos, color='black')
                current_pos += between_outcome_increment
            
            continue
        
        # Adjust y-coordinate to add spacing between outcome groups
        #i*0.1 + j * 0.01 + i * outcome_spacing*0.1
        ax.errorbar(pe, current_pos, xerr=[[xerr_lower], [xerr_upper]], fmt='o', color=colors[j], capsize=5, label=intvn_map[intervention])
        annot = list(data[intervention + '_an'])[i]
        #print(intervention, i, annot, pe)
        if not isinstance(annot, float):
            #ax.annotate(annot, (pe, current_pos),size=15)
            plt.text(pe, current_pos, annot, size=15, color=colors[j])
        current_pos += increment
        
        #print(i, outcome, j, intervention)
        
        if j == (len(interventions) - 1):
            #plt.axhline(current_pos, color='black')
            current_pos += between_outcome_increment

#ax.set_ylim([0,0.5]) 

# Set y-ticks and labels
ax.set_yticks(1.1*np.arange(len(outcomes)))
ax.set_yticklabels(outcomes)

# Add legend
handles, labels = ax.get_legend_handles_labels()
unique_labels = {label: handle for handle, label in zip(handles, labels)}
ax.legend(unique_labels.values(), unique_labels.keys(), loc='lower right')

plt.axvline(0, color='black')

ax.set_xlabel('Point Estimate (Partial Correlation Coefficient)')
ax.set_ylabel('Outcome')

plt.tight_layout()
plt.show()
```
