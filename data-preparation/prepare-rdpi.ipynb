```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
rdpi = pd.read_csv('../data/raw/rdpi/rdpi.csv')
```

```python
rdpi
```

```python
months = []
years = []
for date in rdpi['DATE']:
    this_date = datetime.strptime(date, '%Y-%m-%d')
    months.append(this_date.month)
    years.append(this_date.year)
rdpi['Month'] = months
rdpi['Year'] = years
```

```python
rdpi.to_csv('../data/prepared/rdpi/rdpi.csv')
```
