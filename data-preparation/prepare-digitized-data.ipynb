```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
digitized_data_folder = '../data/prepared/digitized/weekly/'
filenames = {'Zhao': 'Zhao.csv',
'StewartMilk': 'StewartMilk.csv',
'StewartPBMilk': 'StewartPBMilk.csv'}
for filename in filenames:
    filenames[filename] = digitized_data_folder +  filenames[filename]
```

```python
pd.read_csv(filenames['StewartMilk'])
```

```python
merged = None
for outcome in filenames:
    df = pd.read_csv(filenames[outcome]).drop(columns=['Unnamed: 0'])
    if merged is None:
        merged = df
    else:
        merged = merged.merge(df, on=['Year', 'Month', 'Date'], how='outer')
```

```python
merged = merged.sort_values(['Date'])
merged
```

```python
merged.to_csv(digitized_data_folder + 'merged.csv',index=False)
```
